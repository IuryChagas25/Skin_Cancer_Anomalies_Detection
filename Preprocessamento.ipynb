### Funo implementada

```python
def resize_img(dim,directory,formato, future_directory):
    '''Dimenso, Diretrio de origem, formato da imagem, diretrio futuro'''
    i = 1
    for filename in glob.glob(directory + '/*'+ formato):
        img = cv2.imread(filename)
        image = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
        cv2.imwrite('{}{}{}'.format(future_directory +'/',i,'.jpg'),image)
        i+=1
```

```python
resize_img((250,250),'DR_test/DR_testBCC','jpg','DR_test/DR_testBCC_resized')
```

```python
labels = ['MEL', 'NV', 'BCC', 'AKIEC', 'BKL', 'DF', 'VASC']
i = 0
for class_name in labels:
    if not os.path.exists('DR_test/' + str(i) + 'Testemodified'):
        os.mkdir('DR_test/' + str(class_name) + 'Testemodified')
        dr = 'DR_test/' + 'DR_test' + str(class_name)
        dr_future = 'DR_test/' + str(class_name) + 'Testemodified'
        resize_img((250,250), dr, 'jpg', dr_future)
        i+=1
    else:
        i+=1
        continue
```

```python
labels = ['MEL', 'NV', 'BCC', 'AKIEC', 'BKL', 'DF', 'VASC']
i = 0
for class_name in labels:
    if not os.path.exists('DR_train2/' + str(i) + '_resized'):
        os.mkdir('DR_train2/' + str(class_name) + '_resized')
        dr = 'DR_train2/' + str(class_name)
        dr_future = 'DR_train2/' + str(class_name) + '_resized'
        resize_img((250,250), dr, 'jpg', dr_future)
        i+=1
    else:
        i+=1
        continue
```

```python
labels = ['MEL', 'NV', 'BCC', 'AKIEC', 'BKL', 'DF', 'VASC']
i = 0
for class_name in labels:
    if not os.path.exists('DR_validation/' + str(i) + 'Validmodified'):
        os.mkdir('DR_validation/' + str(class_name) + 'Validmodified')
        dr = 'DR_validation/' + str(class_name)
        dr_future = 'DR_validation/' + str(class_name) + 'Validmodified'
        resize_img((250,250), dr, 'jpg', dr_future)
        i+=1
    else:
        i+=1
        continue
```
